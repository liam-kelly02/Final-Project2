```python
#Text-Based Adventure Game
import random

#base player class
class Player:
    def __init__(self, name, health=100, shield=0, damage=25, min_damage = 10):
        self.name = name
        self.health = health
        self.shield = shield
        self.damage = damage
        self.bag = []

    def attack(self, target):
        damage_dealt = random.randint(self.min_damage, self.damage)
        target.take_damage(damage_dealt)
        return damage_dealt

    def take_damage(self, damage):
        damage -= self.shield
        if damage > 0:
            self.health -= damage
            print(f"{self.name} took {damage} damage!")
            print(f"You have {self.health} health remaining")
        else:
            print(f"{self.name} blocked the attack with their shield!")

    def is_alive(self):
        return self.health > 0

    def pickup_potion(self, item):
        print("You landed on a lucky square and found a potion!")
        self.bag.append(item)
        print("You added the potion to your bag\n")

    def use_potion(self):
        if "Potion" in self.bag:
            self.health += 20
            print(f"\n{self.name} used a potion and gained 20 health!")
            print(f"You are now at {self.health} health.\n")
            print("You will now roll the dice")
            self.bag.remove("Potion")
        else:
            print("\nNo potions in the bag!\n")
            print("You will now roll the dice")

#stats for knight class
class Knight(Player):
    def __init__(self, name, health=120, shield=15, damage=20, min_damage = 15):
        super().__init__(name, health, shield, damage, min_damage)
        self.min_damage = min_damage

#stats for barbarian class
class Barbarian(Player):
    def __init__(self, name, health=100, shield=10, damage=30, min_damage = 12):
        super().__init__(name, health, shield, damage, min_damage)
        self.min_damage = min_damage

#stats for assassin class
class Assassin(Player):
    def __init__(self, name, health=75, shield=5, damage=45, min_damage = 5):
        super().__init__(name, health, shield, damage, min_damage)
        self.min_damage = min_damage

#base enemy class
class Enemy:
    def __init__(self, name, health, damage):
        self.name = name
        self.health = health
        self.damage = damage

    def attack(self, target):
        damage_dealt = random.randint(15, self.damage)
        target.take_damage(damage_dealt)
        return damage_dealt

    def take_damage(self, damage):
        self.health -= damage
        if self.health <= 0:
            print(f"{self.name} took {damage} damage!")
        else:
            print(f"{self.name} took {damage} damage!")
            print(f"{self.name} has {self.health} health remaining!")

    def is_alive(self):
        return self.health > 0

#welcome statement
def welcome():
    print("Hello. Welcome!")

#choosing class
def select_character():
    print("Choose your character:")
    print("  1. Knight (Health: 120, Shield: 15, Damage: 20)")
    print("  2. Barbarian (Health: 100, Shield: 10, Damage: 30)")
    print("  3. Assassin (Health: 75, Shield: 5, Damage: 45)")

    choice = input("Enter the number of your choice: ")
    if choice == "1":
        print("\nYou selected the Knight class.")
        print("Good luck!")
        return Knight
    elif choice == "2":
        print("\nYou selected the Barbarian class.")
        print("Good luck!")
        return Barbarian
    elif choice == "3":
        print("\nYou selected the Assassin class.")
        print("Good luck!")
        return Assassin
    else:
        print("\nInvalid choice. Defaulting to Knight.")
        return Knight

#orc fight
def encounter_orc(player):
    print("You have encountered an orc! Prepare to fight.")
    #orc stats
    orc = Enemy("Orc", 50, 40)

    while player.is_alive() and orc.is_alive():
        orc_math1 = random.randint(2, 20)
        orc_math2 = random.randint(2, 15)
        orc_answer = orc_math1 * orc_math2
        print(f"\nYou must answer this question correctly to attack the orc:")
        print(f"What is the product of {orc_math1} and {orc_math2}?")
        player_answer = int(input())
        if player_answer == orc_answer:
            print("\nCorrect! You attacked the Orc.")
            player.attack(orc)
        else:
            print("\nIncorrect! The Orc attacks you.")
            print(f"The correct answer was {orc_answer}.\n")
            orc.attack(player)

    if player.is_alive():
        print("You defeated the orc!\n")
    else:
        print(f"{player.name} was defeated!")

#goblin fight
def encounter_goblin(player):
    print("You have encountered a goblin! Prepare to fight.")
    #goblin stats
    goblin = Enemy("Goblin", 25, 25)

    while player.is_alive() and goblin.is_alive():
        goblin_math1 = random.randint(10, 150)
        goblin_math2 = random.randint(10, 100)
        goblin_answer = goblin_math1 + goblin_math2
        print(f"\nYou must answer this question correctly to attack the goblin:")
        print(f"What is the sum of {goblin_math1} and {goblin_math2}?")
        player_answer = int(input())
        if player_answer == goblin_answer:
            print("\nCorrect! You attacked the goblin.")
            player.attack(goblin)
        else:
            print("\nIncorrect! The goblin attacks you.")
            print(f"The correct answer was {goblin_answer}.\n")
            goblin.attack(player)

    if player.is_alive():
        print("You defeated the goblin!\n")
    else:
        print(f"{player.name} was defeated!")

#dragon fight
def encounter_dragon(player):
    print("You must now fight the dragon!")
    #dragon stats
    dragon = Enemy("Dragon", 100, 50)

    while player.is_alive() and dragon.is_alive():
        dragon_math1 = random.randint(100, 250)
        dragon_math2 = random.randint(50, 150)
        dragon_answer = dragon_math1 - dragon_math2
        print(f"\nYou must answer this question correctly to attack the dragon")
        print(f"What is the difference between {dragon_math1} and {dragon_math2}?")
        player_answer = int(input())
        if player_answer == dragon_answer:
            print("\nCorrect! You attack the dragon")
            player.attack(dragon)
        else:
            print("Incorrect! The dragon attacks you")
            print(f"The correct answer was {dragon_answer}.\n")
            dragon.attack(player)

    if player.is_alive():
        print("You defeated the dragon!\n")
    else:
        print(f"{player.name} was defeated!")

#picking up a sword
def pickup_sword(player):
    print("You landed on a lucky square and found a sword!")
    player.damage += 25
    player.min_damage += 25
    print(f"Your damage increased by 25! Your damage is now {player.damage}.\n")

#picking up a knife
def pickup_knife(player):
    print("You landed on a lucky square and found a knife!")
    player.damage += 10
    player.min_damage += 10
    print(f"Your damage increased by 10! Your damage is now {player.damage}.\n")

#rolling the dice
def roll_dice():
    die1 = random.randint(1,6)
    die2 = random.randint(1,6)
    return die1 + die2

#running main program
def main():
    welcome()

    game_length = 100
    location = 0
    username = input("What is your name? ")

    # Select character
    character_class = select_character()
    player = character_class(username)

    #checks to see if player has reached the end and is still alive
    while location < game_length and player.is_alive():
        turn_input = input("It is your turn to move. What would you like to do? (Enter number)\n1. Roll the dice\n2. Open your bag\n3. View your stats\n")
        if turn_input.strip():
            turn = int(turn_input)
            #player chooses to roll the dice
            if turn == 1:
                roll = roll_dice()
                location += roll
                print(f"\nYou rolled {roll}")
                #player lands on 25, 50, or 75 and gets a sword
                if location == 25 or location == 50 or location == 75:
                    pickup_sword(player)
                #player lands on a multiple of 10 (10, 20, 30, etc.) and gets a knife
                if location % 10 == 0:
                    pickup_knife(player)
                #player lands on a multple of 5 and skips multiples of 10 (5, 15, 25) and gets a potion
                if location % 10 == 5:
                    player.pickup_potion("Potion")
                #if a player rolls a 12 or 2, they have to fight an orc
                if roll == 12 or roll == 2:
                    encounter_orc(player)
                #if a player rolls a 10 or 7, they have to fight a goblin
                if roll == 10 or roll == 7:
                    encounter_goblin(player)
                print(f"You are {game_length - location} spaces away from reaching the end.\n")
            #player opens bag and uses potion
            elif turn == 2:
                print(f"\nYour bag contains: {player.bag}")
                item_use = int(input("What would you like to do?\n1. Use item\n2. Roll the dice\n"))
                if item_use == 1:
                    player.use_potion()
                    roll = roll_dice()
                    location += roll
                    print(f"You rolled {roll}")
                    #player lands on 25, 50, or 75 and gets a sword
                    if location == 25 or location == 50 or location == 75:
                        pickup_sword(player)
                    #player lands on a multiple of 10 (10, 20, 30, etc.) and gets a knife
                    if location % 10 == 0:
                        pickup_knife(player)
                    #player lands on a multple of 5 and skips multiples of 10 (5, 15, 25) and gets a potion
                    if location % 10 == 5:
                        player.pickup_potion("Potion")
                    #if a player rolls a 12 or 2, they have to fight an orc
                    if roll == 12 or roll == 2:
                        encounter_orc(player)
                    #if a player rolls a 10 or 7, they have to fight a goblin
                    if roll == 10 or roll == 7:
                        encounter_goblin(player)
                    print(f"You are {game_length - location} spaces away from reaching the end.")
                #skip item use after opening bag and roll dice
                if item_use == 2:
                    roll = roll_dice()
                    location += roll
                    print(f"\nYou rolled {roll}")
                    #player lands on 25, 50, or 75 and gets a sword
                    if location == 25 or location == 50 or location == 75:
                        pickup_sword(player)
                    #player lands on a multiple of 10 (10, 20, 30, etc.) and gets a knife
                    if location % 10 == 0:
                        pickup_knife(player)
                    #player lands on a multple of 5 and skips multiples of 10 (5, 15, 25) and gets a potion
                    if location % 10 == 5:
                        player.pickup_potion("Potion")
                    #if a player rolls a 12 or 2, they have to fight an orc
                    if roll == 12 or roll == 2:
                        encounter_orc(player)
                    #if a player rolls a 10 or 7, they have to fight a goblin
                    if roll == 10 or roll == 7:
                        encounter_goblin(player)
                    print(f"You are {game_length - location} spaces away from reaching the end.")
            #check stats input
            elif turn == 3:
                print(f"\n{player.name}'s Stats:\n  Health: {player.health}\n  Max Damage: {player.damage}\n  Minimum Damage: {player.min_damage}\n  Shield: {player.shield}")
            else:
                print("Invalid input please try again")
        else:
            print("Please enter a valid input.")

    #player reaches the end and encounters dragon
    if location >= game_length:
        encounter_dragon(player)
        if player.is_alive():
            print("Congratulations you win!")
        else:
            print("Game Over")
    else:
        print("Game Over")

if __name__ == "__main__":
    main()

```
